# Data Analysis

## Payroll

```python
employees['Total Salary'] = employees[['Salario Base','Impostos', 'Beneficios', 'VT', 'VR']].sum(axis=1)
display(employees.head(5))
```

```python
totalPayroll = employees['Total Salary'].sum()
print(f'Monthly payroll is: R$ {totalPayroll:,}')
```
